```python
import pandas as pd

iris = pd.read_csv("iris.csv")
iris
```

```python
iris.groupby(['species']).max()
```

```python
sepal_data = iris.groupby(['species']).agg({'sepal_length': ['mean','max', 'min', 'count']})
```

```python
sepal_data
```

```python
groupings = iris.groupby('species')
versicolor = groupings.get_group('versicolor')
versicolor.max()
```

# reshaping

```python
 df = pd.DataFrame({"Region":['North','West','East','South','North','West','East','South'],
          "Team":['One','One','One','One','Two','Two','Two','Two'],
          "Revenue":[7500,5500,2750,6400,2300,3750,1900,575],
            "Cost":[5200,5100,4400,5300,1250,1300,2100,50]})
df
```

```python
revenue = df.pivot(index="Region", columns="Team", values="Revenue")
revenue
```

```python
df2 = df.set_index(["Region", "Team"])
```

```python
df2 #multiindexed
```

```python
stacked = pd.DataFrame(df2.stack())
stacked
```

```python
stacked.unstack(-3)  #regions in columns
```
